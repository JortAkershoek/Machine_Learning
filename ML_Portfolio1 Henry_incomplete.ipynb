### Opdracht 2
Voor deze opdracht hebben we een dataset met gelabeld data, dus we hebben hier te maken met Supervised Learning.
De outcome bevat 0 of 1, hier hebben we te maken met classification.
Metrics die gebruik kunnen worden zijn: F1 score, recall, precision, accuracy, confusion matrix.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

X = df_train3.drop('Outcome', axis=1)
y = df_train3['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lr = LogisticRegression(max_iter=500, random_state=42)
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
y_pred = lr.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels = ['no diabetes', 'diabetes'])
disp.plot()
plt.show()
```

```python
TN = cm[0,0]
TP = cm[1,1]
FN = cm[1,0]
FP = cm[0,1]

precision = TP / (TP+FP)
recall = TP / (TP+FN) 
F1 = 2 * (precision * recall) / (precision + recall)
print(f"""precision = {precision}
recall = {recall}
f1_score = {F1}""")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
rf = RandomForestClassifier(n_estimators=100, criterion='entropy', max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

recall = recall_score(y_test, y_pred)
print(recall)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

SEED = 42
dt = DecisionTreeClassifier(max_depth=10)
ab = AdaBoostClassifier(base_estimator=dt, n_estimators=100, random_state=SEED)
ab.fit(X_train, y_train)
y_pred = ab.predict(X_test)
recall = recall_score(y_test, y_pred)
print(recall)
```
